```python
import pandas as pd 
import matplotlib.pyplot as plot 
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# To print first five rows of the dataframe
data = pd.read_csv('iris.csv')
df=pd.DataFrame(data)
print(df.head())
```

```python
# To print rows and columns of the dataframe  
print(f'Rows: {df.shape[0]}, Columns: {df.shape[1]}')
```

```python
# To rename all columns with different names
new_col=['feature1','feature2','feature3','feature4','label']
df.columns=new_col
print(df.head())
```

```python
# To extract a class of data in a separate dataframe say ‘y’ 
# and then drop it from the original  data frame. 
# Display the resultant dataframe data and y.  

df_class1 = df[df['label']=='Iris-versicolor']
df.drop(df_class1.index, inplace=True)
print(df_class1.head())
print(df.head())
```

```python
# To print information of the input dataframe ‘data’  
# To print statistical information of the input dataframe ‘data’
df = pd.concat([df,df_class1],ignore_index=True)
print(df.describe())
print(df.info())
```

```python
# df.plot(kind='3Dscatter', x='feature4', y='label')
scatter_data = df.drop(columns=['label'])

pd.plotting.scatter_matrix(scatter_data, figsize=(8, 8), diagonal='kde', alpha=0.7)

plot.suptitle("Scatterplot Matrix of original data")
plot.show()
```

```python
scaler = MinMaxScaler()
df_minmax = pd.DataFrame(scaler.fit_transform(scatter_data), columns=scatter_data.columns)
pd.plotting.scatter_matrix(df_minmax, figsize=(8, 8), alpha=0.7)
plot.suptitle("Scatter matrix of Min-Max Normalized Data")
plot.show()
```
